```python
from bs4 import BeautifulSoup
def clean_html(raw_html):
    if pd.isna(raw_html):
        return ""
    return BeautifulSoup(str(raw_html), "html.parser").get_text(separator=" ", strip=True)

df["description"] = df["description"].apply(clean_html)
df.columns, df["description"].iloc[0]
```

```python
import os
import pandas as pd
import requests
from tqdm import tqdm
from PIL import Image
from io import BytesIO
import torch
from transformers import CLIPProcessor, CLIPModel
import pickle
import math
```

```python
HF_REPO_BASE = "https://huggingface.co/datasets/Sharathsoorya/Multimodal_Fashion_Data/resolve/main/"
LOCAL_IMG_DIR = "./hf_images"
OUTPUT_FILE = "fashion_embeddings.pkl"
BATCH_SIZE = 64  # batch size for Qdrant upload
MAX_CHUNK_LEN = 65  # CLIP token limit

COLS_TO_STORE = ['gender_x', 'masterCategory_x', 'subCategory_x', 'articleType_x',
                 'baseColour_x', 'season_x', 'year_x', 'usage_x', 'productDisplayName_x',
                 'brandName', 'articleAttributes', 'description', 'price',
                 'discountedPrice', 'rating', 'image_path']

os.makedirs(LOCAL_IMG_DIR, exist_ok=True)
```

```python
device = "cuda" if torch.cuda.is_available() else "cpu"
print(f'Using : {device}')
clip_model = CLIPModel.from_pretrained("openai/clip-vit-base-patch32").to(device)
clip_processor = CLIPProcessor.from_pretrained("openai/clip-vit-base-patch32")
```

```python
def download_image_from_hf(image_path):
    local_path = os.path.join(LOCAL_IMG_DIR, os.path.basename(image_path))
    if not os.path.exists(local_path):
        # take only the filename from image_path
        filename = os.path.basename(image_path)
        url = HF_REPO_BASE + filename
        r = requests.get(url)
        if r.status_code == 200:
            img = Image.open(BytesIO(r.content)).convert("RGB")
            img.save(local_path)
        else:
            print(f"Failed to download {image_path}: {r.status_code}")
            return None
    return local_path
```

```python
def chunk_text(text, max_len=MAX_CHUNK_LEN):
    """
    Splits text into chunks of max_len tokens.
    """
    tokens = clip_processor.tokenizer.encode(text)
    chunks = []
    for i in range(0, len(tokens), max_len):
        chunk_tokens = tokens[i:i+max_len]
        chunk_text = clip_processor.tokenizer.decode(chunk_tokens)
        chunks.append(chunk_text)
    return chunks
```

```python
data_to_save = []

for idx, row in tqdm(df.iterrows(), total=len(df)):
    try:
        img_path = download_image_from_hf(row['image_path'])
        if img_path is None:
            continue
        text_chunks = chunk_text(row['description'])

        chunk_embeddings = []
        for chunk in text_chunks:
            inputs = clip_processor(
                text=chunk,
                images=Image.open(img_path).convert("RGB"),
                return_tensors="pt",
                truncation=True,
                padding=True
            ).to(device)

            with torch.no_grad():
                text_emb = clip_model.get_text_features(inputs['input_ids'], inputs['attention_mask'])
                img_emb = clip_model.get_image_features(inputs['pixel_values'])
                combined_emb = torch.cat([text_emb, img_emb], dim=-1)  # concat text + image
                chunk_embeddings.append(combined_emb)

        final_emb = torch.mean(torch.stack(chunk_embeddings), dim=0).cpu().numpy()[0]

        payload = {col: row[col] for col in COLS_TO_STORE}

        data_to_save.append({
            "id": int(idx),
            "vector": final_emb.tolist(),
            "payload": payload
        })


    except Exception as e:
        print(f"Error processing row {idx}: {e}")

# Save remaining embeddings
if data_to_save:
    with open(OUTPUT_FILE, "wb") as f:
        pickle.dump(data_to_save, f)

print(f"All embeddings and payloads saved to {OUTPUT_FILE}")
```
